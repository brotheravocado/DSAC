```python
html = requests.get("http://www.naver.com").text
soup = BeautifulSoup(html)
soup
```

## More on BeautifulSoup

```python
from bs4 import BeautifulSoup
```

```python
html_doc = """
<html><body>
<h1> 파이썬으로 웹문서읽기</h1>
<p> 페이지 분석기능</p>
<p> 페이지 정렬</p>
</body>
</html>
"""
```

```python
soup = BeautifulSoup(html_doc,"html.parser")
```

```python
h1= soup.html.body.h1; h1
```

```python
h1= soup.html.body.h1; h1.string
```

```python
p1 = soup.html.body.p ; p1
```

```python
soup.html.body.p.next_sibling # 리턴만 나옴
```

```python
p2 = soup.html.body.p.next_sibling.next_sibling ; p2
```

```python
print("h1 = "+ h1.string)
print("p1 = "+ p1.string)
print("p2 = "+ p2.string)
```

## id를 이용하는 방법

```python
html_doc = """
<html><body>
<h1 id= "title"> 파이썬으로 웹문서읽기</h1>
<p id = "body1"> 페이지 분석기능</p>
<p id = "body2"> 페이지 정렬</p>
</body>
</html>
"""
```

```python
soup = BeautifulSoup(html_doc,"html.parser")
```

```python
title = soup.find(id="title"); title
```

```python
body1 = soup.find(id="body1"); body1
```

```python
body2 = soup.find(id="body2"); body2
```

```python
print("title = "+ title.string)
print("body1 = "+ body1.string)
print("body2 = "+ body2.string)
```

```python
## find_all()를 사용하는 경우
```

```python
html_doc = """
<html><body>
 <ul>
    <li><a href = "http://www.naver.com">naver</a?></li>
        <li><a href = "http://www.daum.net">daum</a?></li>
 <ul>
</body></html>
"""
```

```python
soup = BeautifulSoup(html_doc,"html.parser")
links = soup.find_all("a")
```

```python
for aa in links:
    href = aa.attrs["href"]
    text = aa.string
    print(text,"--->",href)
```
