# 스타벅스 매장 데이터를 가지고 와서 지도에 표기


```python
import requests, json
from pandas.io.json import json_normalize
```

```python

data = {
    'ins_lat':'37.56682',
    'ins_lng':'126.97865',
    'p_sido_cd':'05',# 01 - 서울, 05 - 부산?
    'p_gugun_cd':'', # 없으면 전체
    'in_biz_cd':'',
    'set_date':'',
    'iend':'1000'
}
```

```python
url = 'https://www.istarbucks.co.kr/store/getStore.do'
r = requests.post(url,data=data)
```

```python
r.text
```

```python
jo = json.loads(r.text) ; jo # JSON decode
```

```python
df = json_normalize(jo,'list'); df
```

```python
len(df)
```

```python
df.columns
```

```python
## 필요한 컬럼만 추출
col = ['s_name','lat','lot','sido_name','gugun_name','doro_address','tel']
df = df[col]
```

```python
df
```

```python
df.head()
```

```python
df.dtypes
```

```python
# 위도, 경도 정보의 데이터 타입 변환
df['lat'] = df['lat'].astype(float)
df['lot'] = df['lot'].astype(float)
df.dtypes
```

```python
# 특정지점 확인
df[df['s_name']=='부산대역']
```
